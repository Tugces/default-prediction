##### SMOTE 

```python
X = dfTestTrain_concat.iloc[:,2:]
y = dfTestTrain_concat['default']

# transform the dataset
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

print('X Sizes: ', X.shape)
print('y Sizes: ', y.shape)
print('X_smote Sizes: ', X_smote.shape)
print('y_smote Sizes: ', y_smote.shape)
```

##### Train-Test Split 

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size=0.30, random_state=42)

print('X_train Sizes: ', X_train.shape)
print('X_test Sizes: ', X_test.shape)
print('y_train Sizes: ', y_train.shape)
print('y_test Sizes: ', y_test.shape)
```

#### Model Development 

##### Baseline Model - Logistic Regression

```python
model_name='base_model'
scaler = StandardScaler()
lr = LogisticRegression()
pipeline = Pipeline(steps=[('scaler', scaler),
                            ('model', lr)])
metrics_dict, pred_train, pred_test, preds_proba = run_model_metrics.run_pipeline(model_name, pipeline, 
                                                                                  X_train, y_train, 
                                                                                  X_test, y_test)
df_metrics = pd.DataFrame([metrics_dict])
df_metrics
```

```python
# according to the baseline model, feature_importance
# get importance
importances = pd.DataFrame(data={
    'feature': X_train.columns,
    'importance': lr.coef_[0]
})
lr_importances = importances.sort_values(by='importance', ascending=False)
lr_importances
```

```python
# For the correlated features, feature importance will checked
corr_columns
```

For the correlated features, feature importance will checked.

It was decided that first features will be used if the score is bigger than 0.1 or less than -0.1.
If the correlated features are in this range, then remove the one with the less importance level.

According to the evaluation, below features should be removed after the range filter.
'max_paid_inv_0_12m',
'status_max_archived_0_6_months',
'status_2nd_last_archived_0_24m'


```python
# according to the range get the list of features
keep_list = list(lr_importances[(lr_importances['importance'] >= 0.1) | (lr_importances['importance'] <= -0.1)]['feature'])
# features in the keep_list needs to be removed because of the correlation
remove_list =['max_paid_inv_0_12m', 'status_max_archived_0_6_months', 'status_2nd_last_archived_0_24m']
# remove correlated
keep_list_upd = [feature for feature in keep_list if feature not in remove_list]
keep_list_upd
```

##### GridSearch for Base Model 

```python
scaler = StandardScaler()
lr = LogisticRegression()
pipeline = Pipeline(steps=[('scaler', scaler),
                            ('classifier', lr)])

param_grid = {"classifier__C": [0.001, 0.01, 0.1, 1, 10],
              "classifier__max_iter" : [10,15]
             }
grid = GridSearchCV(pipeline, param_grid, cv=5, scoring = 'f1') 
grid.fit(X_train[keep_list_upd], y_train)
print("Train Data-Best cross-validation score: {:.2f}".format(grid.best_score_)) 
print("Best parameters: ", grid.best_params_)
print("Test Data-Grid Search Results: {:.2f}".format(grid.score(X_test[keep_list_upd], y_test)))  

```
